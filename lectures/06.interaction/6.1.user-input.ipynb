## Yes/no choice

```python
# convention: capital for the default behaviour
user_text = input("Do you want to continue? [N/y]").lower()

# use a loop to keep prompting the user
while True:
    # if user_text is empty, defaults as no
    if not user_text or user_text in ["n", "no"]:
        print("The programme stops.")
        break
    elif user_text in ["y", "yes"]:
        print("The programme continues.")
        break
    else:
        user_text = input("Answer not recognised, please try again. [Y/n]").lower()
```

## Pick an option

```python
options = [
    "Countdown to anarchy initiated.",
    "Our team will start laying out the foundations at once.",
    "The world has been reset. Enjoy the big bang!"
]

user_text = input("""Options:
1 = Breed chaos.
2 = Build a structure.
3 = Tabula rasa.
Which option do you choose? [Please write only the number!]\n""")

# use a loop to keep prompting the user
while True:
    try:
        # convert from human counting (from 1) to programming (from 0)
        first_letter = user_text[0]
        user_text = int(first_letter) - 1
        if user_text >= 0 and user_text < len(options):
            # use the response as index to retrieve the option
            choice = options[user_text]
            break
        else:
            user_text = input("Only three options available, please try again. [Please write only the number!]\n").lower()
    except:
        user_text = input("Answer not recognised, please try again. [Please write only the number!]\n").lower()

print("-" * 40)
print(choice)
```

## Pattern matching

[doc](https://docs.python.org/3/reference/compound_stmts.html#match)  
[tutorial](https://docs.python.org/3/tutorial/controlflow.html#match-statements), [RealPython](https://realpython.com/structural-pattern-matching/)

```python
user_text = input("Please choose your character: ").lower().strip()

match user_text:
    case "iris":
        print("Hello Iris.")
    case "jeremie":
        print("Hello, Jérémie.")
    case _:
        print("Hello, stranger.")
        
print("Let the game begin!")
```
